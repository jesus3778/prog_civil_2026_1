# **Dosificación de Mezcla para Columnas**

**planteamiento**

**Un** ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular. Se sabe que para un concreto estándar de 210 kg/cm2
 (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

Requerimientos:

1. Solicitar la altura de la columna en metros.
2. Solicitar el diámetro de la columna en metros.
3. Calcular el volumen de la columna: $V=\pi*r^2*h$
. (Sugerencia: usar 3.1416
 o importar math.pi).
4. Calcular la cantidad de sacos: Sacos=$V*9.5$
.

**Prueba del programa:**

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2):

Volumen ≈0.377 m3.
Sacos necesarios ≈3.58
 (el estudiante verá aquí la importancia de los tipos decimales).

````python
#entrada: Solicitar la altura de la columna en metros.
#Solicitar el diámetro de la columna en metros.
h = float(input("ingrese la altura de la columna en metros"))
d = float(input("ingrese el diametro de la columna en metros"))
#proceso:calcula el radio r,calcula el volumen de la columna,calcula sacos
r = d/2
V = 3.1416*r**2*h
Sacos = V*9.5
print("el volumen de la columna es",V)
print("la cantidad de sacos es",Sacos)
````

Aquí está la explicación de cada línea del código para el cálculo de sacos de cemento:

```python
h = float(input("ingrese la altura de la columna en metros"))
```
Esta línea solicita al usuario que introduzca la altura de la columna en metros. La función `input()` lee la entrada como una cadena de texto, y `float()` convierte esa cadena en un número decimal (flotante) para que pueda usarse en cálculos.

```python
d = float(input("ingrese el diametro de la columna en metros"))
```
Similar a la línea anterior, esta línea solicita al usuario el diámetro de la columna en metros y lo convierte a un número decimal.

```python
r = d/2
```
Esta línea calcula el radio (`r`) de la columna dividiendo el diámetro (`d`) introducido por el usuario entre 2.

```python
V = 3.1416*r**2*h
```
Esta línea calcula el volumen (`V`) de la columna cilíndrica utilizando la fórmula $V = \pi * r^2 * h$. Aquí, `3.1416` se usa como una aproximación de Pi, `r**2` calcula el radio al cuadrado, y `h` es la altura.

```python
Sacos = V*9.5
```
Esta línea calcula la cantidad total de sacos de cemento (`Sacos`) necesarios. Multiplica el volumen calculado de la columna (`V`) por la cantidad de sacos de cemento requeridos por metro cúbico (9.5).

```python
print("el volumen de la columna es",V)
```
Esta línea imprime en la consola el resultado del volumen calculado de la columna, junto con un mensaje descriptivo.

```python
print("la cantidad de sacos es",Sacos)
```
Finalmente, esta línea imprime la cantidad total de sacos de cemento calculados, junto con un mensaje descriptivo.
