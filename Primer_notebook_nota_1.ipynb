# 1. Problema de ingeniería (ejemplo sencillo)

Un ingeniero agrícola necesita calcular **el área de un terreno rectangular** para estimar cuánta semilla debe sembrar.

La fórmula del área es:

A = b * h

donde:

* **A** = área del terreno (m²)
* **b** = base del terreno (m)
* **h** = altura o largo (m)

---

# 2. Definición del algoritmo (Entradas – Proceso – Salidas)

### Entradas

Datos que el usuario debe ingresar.

* base del terreno (b)
* altura del terreno (h)

### Proceso

Operación matemática que resuelve el problema.

* calcular el área
* área = base × altura

### Salida

Resultado que el programa mostrará.

* área del terreno en metros cuadrados

---

# 3. Diseño del algoritmo (explicación lógica)

1. Pedir al usuario la **base del terreno**.
2. Pedir al usuario la **altura del terreno**.
3. Multiplicar base por altura.
4. Mostrar el resultado del área.

---

# 4. Pseudocódigo

```
Inicio

    Escribir "Ingrese la base del terreno"
    Leer base

    Escribir "Ingrese la altura del terreno"
    Leer altura

    area = base * altura

    Escribir "El área del terreno es:", area

Fin
```

---

# 5. Código en Python

```python
# Programa para calcular el área de un terreno rectangular
# Aplicación básica para ingeniería agrícola o civil

# Entrada de datos
base = float(input("Ingrese la base del terreno en metros: "))
altura = float(input("Ingrese la altura del terreno en metros: "))

# Proceso
area = base * altura

# Salida
print("El área del terreno es:", area, "m²")
```
